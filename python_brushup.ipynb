## 辞書
キーとそれに対応する要素がセットになったもの。

```python
#キーとそれに対応する要素がセット

dic = {'a':100, 'b':200}
print(dic['a'])
```

```python
# 空の辞書
empty_dic = {}
```

```python
#キーの検索
'a' in dic
```

```python
#キーと要素の追加
dic['c'] = 300
dic
```

```python
#keysメソッド
for k in dic.keys():
    print(k, dic[k])
```

```python
#itemsメソッド
for k, v in dic.items():
    print(k, v)
```

```python
# リストからデータフレームを作る
df = pd.DataFrame([[10,'a'],[20,'b'],[30,'c'],[40,'d']])
print(df)
```

```python
# pd.read_csvでcsvファイルを読み込むこともできる
URL = '/content/drive/MyDrive/projects/python_data_analysis/SSDSE-E-2025.csv'

df = pd.read_csv(URL, skiprows=[0,1,3],
                 encoding='shift_jis')
df
```

```python
# インデックスの情報確認
df.index
```

```python
#カラムの情報確認
df.columns
```

```python
#行の抽出
df[44:47]
```

```python
#列の抽出
#複数の場合は列名のリストで指定する
df[['都道府県','一戸建住宅数']]
```

```python
#条件抽出
df[(df['総人口'] <= 1500000) & (df['出生数'] >= 10000)]
```

```python
#行と列両方任意のものを抽出するには、loc属性を使う
df.loc[[12,26],['都道府県','総人口','出生数']]
```
