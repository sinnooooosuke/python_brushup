## for文色々

```python
# 基本はリストから
lst = [1,2,3,4,5]
s = 0
for x in lst:
  s = s + x
print(s)
```

```python
# enumerateを使うとインデックスも参照できる
for i, x in enumerate(lst):
  print(i,x)
```

```python
# rangeを使うと連続した整数列を代入できる
n=5
s=0
for x in range(1, n+1):
  s = s+x
print(s)
```

## Numpy

配列（ndarray）というデータ構造を提供。
Pandasより処理が速い。

```python
# 配列の属性
array_2d = np.array([[1,2,3],[4,5,6]])
print('次元数：', array_2d.ndim)
print('形状：', array_2d.shape)
```

```python
# 形状の変換
array_2d.reshape(1,6)
```

```python
# -1で自動で補うことも可能
array_2d.reshape(3,-1)
```

```python
# zeros(onesも同様)
np.zeros(5)
```

```python
np.zeros((3,3))
```

```python
# インデックス参照

# ファンシーインデックス参照
# コピーが返るので代入しても元の配列には変化なし
# 第一引数は列のインデックス、第二引数は行のインデックスのリスト。
print(array_2d[[1,0],[1,0]])
```

```python
# ブールインデックス参照
# 以下の式は真偽値からなる配列を返す
array_2d < 3
```

```python
# 上の配列を添え字に指定すると、条件を満たすものだけ抜き出せる。
array_2d[array_2d < 3]
```

```python
# 算術演算も可能

2*array_2d
```

```python
array_2d * array_2d
```

```python
# 勝手に形状の違いを補ってくれる
array_2d * np.array([1,2,3])
```

```python
# ユニバーサル関数
# 各要素に対し同じ処理を行える

np.power(array_2d, 3)
```

# 実用的な事項

```python
URL = '/content/drive/MyDrive/projects/python_data_analysis/SSDSE-E-2025.csv'

df = pd.read_csv(URL, skiprows=[0,1,3],
                 encoding='shift_jis')
```

```python
# ヒストグラム
plt.hist(df['合計特殊出生率'])
plt.xlabel('合計特殊出生率')
plt.ylabel('度数')
plt.show()
```

```python
# 記述統計量
df['合計特殊出生率'].describe()
```

```python
# 箱ひげ図
plt.boxplot(df['合計特殊出生率'], labels=['出生数'])
plt.show()
```

```python
# 列の作成
df['保育所等数'] = df['保育所等数'] * 10000 / df['総人口']
```

```python
# 散布図
plt.scatter(df['合計特殊出生率'], df['保育所等数'])
plt.axvline(df['合計特殊出生率'].mean(), color='gray', linestyle='dashed')
plt.axhline(df['保育所等数'].mean(),color='gray', linestyle='dashed')
plt.xlabel('出生数')
plt.ylabel('保育所等数（1万人あたり）')
plt.show()
```

```python
# 相関係数
df[['合計特殊出生率', '保育所等数']].corr()
```

```python
URL = '/content/drive/MyDrive/projects/python_data_analysis/SSDSE-A-2025.csv'
df = pd.read_csv(URL, skiprows=[0,1],
                 encoding='shift_jis')
df = df[['都道府県','市区町村','総人口','65歳以上人口']]
```

```python
# データの並び替え
df.sort_values('総人口', ascending=False)
```

```python
# 行の条件抽出
df[(df['総人口']==0) | (df['65歳以上人口']==0)]
```

```python
# 欠損値の処理
df.loc[df['総人口']==0, '総人口'] = None
df.dropna(inplace=True)
```

```python
# 外れ値の処理
df['高齢化率'] = df['65歳以上人口'] * 100 / df['総人口']
df.sort_values(['高齢化率'], ascending=False)
```

```python
plt.boxplot(df['高齢化率'], labels=['高齢化率'])
plt.show()
```
